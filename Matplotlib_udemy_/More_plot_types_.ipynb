# More Plot Types

### plt.plot(args) : for line plots
### plt.bar(args) : for bar plots
### plt.scatter(args) : for scatter plots

```python
"""
                    More Plot Types:
--> plt.plot(args) : for line plots
--> plt.bar(args) : for bar plots
--> plt.scatter(args) : for scatter plots

                        ADDING HORIZONTAL AND VERTICAL LINES TO AXES
--> It is sometimes useful to add vertical/horizontal lines to a chart.
    --> We can display info such as mean, median and other important values
--> ax.axhline(y=, xmin=0, xmax=1) : for horizontal lines
    --> xmin/xmax : are values between 0 and 1 and have nothing to do with actual values on the plot.
        --> 0 indicates left edge
        --> 1 indicates right edge
--> ax.axvline(x=, ymin=0, ymax=1) : for vertical lines
    --> ymin/ymax : are values between 0 and 1 and have nothing to do with actual values on the plot.
        --> 0 indicates bottom edge
        --> 1 indicates top edge
        
                        HISTOGRAMS
--> We can use NumPy to generate a histogram data, and then use Matplotlib to generate the bar chart.
--> We can also use built-in Matplotlib directly.
    --> ax.hist(data, bins=)
        --> data is the data for which we want to generate a histogram
        --> bins specifies the number of bins we want to use.
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv('populations.csv')
df.head()
```

```python
df.info()
```

```python
data = df.drop(index=0, 
               columns=['April 1, 2000 Population Estimates Base', 'July 1, 2001 Estimate'])
```

```python
data.columns = ['area', 'population_estimate']
print(data)
```

```python
data = data.sort_values(by='population_estimate', ascending=False)
data.head()
```

```python
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = [10, 8]
fig, ax = plt.subplots()
ax.plot(data['area'].values, 
        data['population_estimate'].values, 
        color='purple', 
        linewidth=0.9, 
        label='Population')
ax.set_xlabel('state')
ax.set_ylabel('population')
ax.legend()
None
```

```python
fig, ax = plt.subplots()
ax.plot(data['area'].values,
        data['population_estimate'].values,
        color='purple',
        linewidth=0.9,
        label='Population')
ax.set_xlabel('state')
ax.set_ylabel('population')
ax.legend()
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=9)
plt.minorticks_on()
None
```

```python
mpl.style.use('seaborn-whitegrid')
fig, ax = plt.subplots()
ax.bar(data['area'].values,
        data['population_estimate'].values,
        color='purple',
        linewidth=0.9,
        label='Population')
ax.set_xlabel('state')
ax.set_ylabel('population')
ax.legend()
plt.xticks(rotation=90, fontsize=9)
plt.yticks(fontsize=9)
plt.minorticks_on()
None
```

```python
import random
random.seed(0)
data_1 = [random.gauss(0, 1) for _ in range(1_000_000)]
```

```python
np.random.seed(0)
data_2 = np.random.normal(0, 1, size=1_000_000)
```

```python
freq1, bins1 = np.histogram(data_1, bins=50)
freq2, bins2 = np.histogram(data_2, bins=50)
```

```python
print(bins1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.bar(bins1[:-1], freq1)
ax1.set_title('Python Generated')
ax2.bar(bins2[:-1], freq2)
ax2.set_title('Numpy Generated')
None
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(bins1[:-1], freq1)
ax1.set_title('Python Generated')
ax2.scatter(bins2[:-1], freq2)
ax2.set_title('Numpy Generated')
None
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(bins1[:-1], freq1)
ax1.set_title('Python Generated')
ax2.plot(bins2[:-1], freq2)
ax2.set_title('Numpy Generated')
None
```

```python
tenth1, fiftieth1, ninetieth1 = np.percentile(data_1, [10, 50, 90])
tenth2, fiftieth2, ninetieth2 = np.percentile(data_2, [10, 50, 90])
fig, ax = plt.subplots()
ax.scatter(bins1[:-1], freq1, label='random')
ax.axvline(tenth1, color='g', linestyle=':', label='10%')
ax.axvline(fiftieth1, color='r', linestyle=':', label='50%')
ax.axvline(
    ninetieth1, color=mpl.colors.to_rgba((0.1, 0.9, 0.1, 1.0)), 
    linestyle=':', label='90%')
ax.legend()
None
```
