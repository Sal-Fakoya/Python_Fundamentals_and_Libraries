```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
"""
                    MATPLOTLIB LIBRARY
--> Matplotlib is a popular graphing library.
--> There are many other available too:
    --> geoplotlib is used for maps and geographical data
    --> ggplot is used in R, simpler than matplotlib, not as customizable (based on matplotlib).
    --> plotly is used for making interactive plots, web and contour plots, 3D-plots, etc.
--> Numerous extension packages to Matplotlib:
    --> Financial
    --> Maps and map projections
    --> Speciality axes (broken axes)
    --> electronic circuits
    --> venn diagrams
    --> density maps
    --> statistical maps
    --> ML visualisations.
--> We can create single plots, overlayed plots, grids of plots.

                        MATPLOTLIB STYLES:
--> mpl.style.available : returns a list of various matplotlib styles available on your system
--> mpl.style.use('string_of_style_from_list')

                        ANATOMY OF FIGURES
--> A figure is a container that contains the plots.
--> axes : are inside the figure. There can be multiple axes. For 2D, x and y axes.
--> major ticks : are on each axes that can be changed or modified 
--> minor ticks : are in between the major ticks.
--> major tick labels : are labels associated with the major ticks on the axes.
--> minor tick labels : are labels associated with the minor ticks on the axes.
--> x-axis label, y-axes labels : are labels associated with x-axes and y-axis label.
--> plot : the plot is the line themselves.
--> markers 
--> legend 
--> title

                    CREATING A FIGURE AND AXES: plt.subplots() in pyplot module.
--> plt.subplots() : creates a new figure and one Axes object.
    --> returns a tuple which contains the figure and the axes.
                fig, ax = plt.subplots()
        --> Here, the figure will be blank because we need to specify something to plot.
        
                        PLOTTING DATA: ax.plot(x_coords, y_coords, label="some_label")
--> Add a plot to an Axes:
--> ax.plot(x_coords, y_coords, label="some_label")
    --> This add a (line) plot to the Axes object, with specified plot name (used in legend)
    --> x_coords and y_coords can be lists, NumpyArrays, Pandas Columns, etc
    --> You can keep adding more plots to the same Axes
    --> We have to display the figure to see the result.
        --> We typcially create figures and plots inside a single cell on Jupyter Notebook.
    
                        ADDING AXES TO THE FIGURE:
--> ax.set_xlabel(some_str) : sets the x-axis label
--> ax.set_ylabel(some_str) : sets the y-axis label
--> ax.set_title(some_str) : sets the title
--> ax.legend() : creates and adds the legend
    --> NOTE THAT all of these are applied to the Axes object.
"""
```

```python
fig, ax = plt.subplots()
x_pts = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y_pts = np.sin(x_pts)
ax.plot(x_pts, y_pts, label="sin")
```

```python
ax.plot(x_pts, np.cos(x_pts), label='cos')
fig
```

```python
ax.set_xlabel('x-axis label')
ax.set_ylabel('y-axis label')
fig
```

```python
ax.set_title('Trig Functions')
ax.legend()
```
